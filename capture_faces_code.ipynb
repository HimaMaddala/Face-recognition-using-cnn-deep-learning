```python
import os
import cv2

def generate_dataset():
    face_classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
    
    def face_cropped(img):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    
        if len(faces) == 0:
            return None
    
        # Get the largest face detected
        largest_face = max(faces, key=lambda face: face[2] * face[3])
        x, y, w, h = largest_face
    
        # Extend the bounding box by a factor (e.g., 1.5)
        extend_factor = 1.5
        x_extended = max(int(x - (extend_factor - 1) * w / 2), 0)
        y_extended = max(int(y - (extend_factor - 1) * h / 2), 0)
        w_extended = min(int(w * extend_factor), img.shape[1] - x_extended)
        h_extended = min(int(h * extend_factor), img.shape[0] - y_extended)
    
        cropped_face = img[y_extended:y_extended+h_extended, x_extended:x_extended+w_extended]
        return cropped_face

    
    cap = cv2.VideoCapture(0)
    img_id = 0
    
    # Specify the directory where images will be saved
    save_dir = "C:/Users/madda/Downloads/face recognition cnn/data"
    os.makedirs(save_dir, exist_ok=True)  # Create the directory if it doesn't exist
    
    while True:
        ret, frame = cap.read()
        if face_cropped(frame) is not None:
            img_id += 1
            face = cv2.resize(face_cropped(frame), (200, 200))
            face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
            
            # Construct the full file path
            file_name_path = os.path.join(save_dir, f"Tony{img_id}.jpg")
            
            cv2.imwrite(file_name_path, face)
            cv2.putText(face, str(img_id), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
            
            cv2.imshow("Cropped_Face", face)
            if cv2.waitKey(1) == 13 or int(img_id) == 200:
                break
    
    cap.release()
    cv2.destroyAllWindows()
    print("Collecting samples is completed !!!")

generate_dataset()
```
